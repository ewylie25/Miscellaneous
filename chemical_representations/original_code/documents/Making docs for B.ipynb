```python
with open('new.dat') as f:
    new = pickle.load(f)
```

```python
new
```

```python
data = []
for j in new:
    d = new[j]
    base = np.zeros((311,311))
    for k,v in d.items():
        i,j = k.split('.')
        base[int(i),int(j)]=v
    data.append(base)
```
